```python
import PyPDF2
from pdf2image import convert_from_path

def extract_page_as_image(pdf_path, page_number, output_image_path):
    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)

        # Get the specified page
        page = pdf_reader.pages[page_number]

        # Convert the page to an image
        images = convert_from_path(pdf_path, first_page=page_number+1, last_page=page_number+1)

        # Save the image to the specified output path
        images[0].save(output_image_path, 'JPEG')

# Example usage
pdf_path = '/content/fluid notes.pdf'  # Path to the PDF file
page_number = 0  # Page number to extract (0-indexed)
output_image_path = 'page_1.jpg'  # Output image path

extract_page_as_image(pdf_path, page_number, output_image_path)
```

**Bounding Box for images, graphs**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
paragraph_coordinates_image=[]
# Read the image
image = cv2.imread('/content/image_extract.png')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Use adaptive thresholding to binarize the image
threshold_image = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)

# Find contours of figures
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Find contours of text regions (assuming text is in black)
_, threshold_text = cv2.threshold(blurred_image, 127, 255, cv2.THRESH_BINARY_INV)
contours_text, _ = cv2.findContours(threshold_text, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Set initial padding sizes for bounding box
padding_plus_x = 90  # Adjust this value to change the initial padding size in the positive x-axis direction
padding_minus_x = 35  # Adjust this value to change the initial padding size in the negative x-axis direction
padding_plus_y = 10  # Adjust this value to change the initial padding size in the positive y-axis direction
padding_minus_y = 25  # Adjust this value to change the initial padding size in the negative y-axis direction

# Scale down padding values if the image is smaller
height, width = image.shape[:2]
if width < 500:
    padding_plus_x = int(padding_plus_x * width / 500)
    padding_minus_x = int(padding_minus_x * width / 500)
    padding_plus_y = int(padding_plus_y * width / 500)
    padding_minus_y = int(padding_minus_y * width / 500)

# Iterate through contours and draw rectangles around non-text contours
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)

    # Ignore contours that are too small (likely noise)
    if cv2.contourArea(contour) < 1000:
        continue

    # Calculate aspect ratio to differentiate between text and non-text regions
    aspect_ratio = w / h

    # Ignore regions with high aspect ratio (likely text)
    if aspect_ratio > 3:
        continue

    # Initialize variables to track the bounding box with initial padding
    x_min = x - padding_minus_x
    x_max = x + w + padding_plus_x
    y_min = y - padding_minus_y
    y_max = y + h + padding_plus_y

    # Check if there are nearby text contours
    if len(contours_text) > 0:
        # Calculate the distance of the nearest text contour
        distances = [cv2.pointPolygonTest(contours_text[i], (x + w // 2, y + h // 2), True) for i in range(len(contours_text))]
        nearest_text_distance = min(distances)

        # Calculate new padding based on the distance of the nearest text contour
        padding_plus_x = int(max(90 - nearest_text_distance * 0.003, 10))
        padding_minus_x = int(max(35 - nearest_text_distance * 0.003, 10))
        padding_plus_y = int(max(10 - nearest_text_distance * 0.0045, 10))
        padding_minus_y = int(max(25 - nearest_text_distance * 0.0045, 10))

        # Adjust the bounding box based on the new padding
        x_min = max(0, x - padding_minus_x)
        y_min = max(0, y - padding_minus_y)
        x_max = min(width, x + w + padding_plus_x)
        y_max = min(height, y + h + padding_plus_y)

    # Draw bounding box around the figure
    paragraph_coordinates_image.append((x_min,y_min,x_max,y_max))
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

# Display the result
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import re

def combine_text_and_math(textOCR, MathOCR):
    # Split textOCR into text segments and math segments
    text_segments = re.split(r'\\mathsf{.+?}', textOCR)
    math_segments = re.findall(r'\\mathsf{.+?}', textOCR)

    combined_latex = ""
    math_index = 0

    # Combine text and math segments
    for text_segment, math_segment in zip(text_segments, math_segments):
        combined_latex += text_segment
        combined_latex += math_segment
        math_index += 1

    # Append any remaining math expressions from MathOCR
    combined_latex += MathOCR[math_index:]

    return combined_latex
```

```python
def add_padding(image, padding_size,color=(255, 255, 255)):
    image_pil = Image.fromarray(image)

    # Calculate the new size
    new_width = image_pil.width + 2 * padding_size
    new_height = image_pil.height + 2 * padding_size

    # Create a new image with the calculated size and fill it with the specified color
    padded_image = Image.new(image_pil.mode, (new_width, new_height), color)

    # Paste the original image onto the padded image
    padded_image.paste(image_pil, (padding_size, padding_size))

    return padded_image
```

```python
def crop_image(image, crop_size):
    height, width = image.shape[:2]
    cropped_image = image[crop_size:height-crop_size, crop_size:width-crop_size]
    return cropped_image
```

```python
def generate_latex(input_string):
    # Split the string into parts
    parts = input_string.split(' ')

    # Construct the LaTeX expression with \text{} around each part
    latex_expression = r'\\text{' + r'}~\\text{'.join(parts) + '}'

    return latex_expression

# Example usage
input_string = "ratio & of the permeability 4 and permittivity ¢ is €"
latex_output = generate_latex(input_string)
print(latex_output)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch


#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)


# Sentences we want sentence embeddings for
sentences = ['This is an example sentence', 'Each sentence is converted']

# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
model = AutoModel.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')

# Tokenize sentences
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# Compute token embeddings
with torch.no_grad():
    model_output = model(**encoded_input)

# Perform pooling. In this case, max pooling.
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

print("Sentence embeddings:")
print(sentence_embeddings)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

tokenizer = Tokenizer()
tokenizer.fit_on_texts(documents_df.documents_cleaned)
tokenized_documents = tokenizer.texts_to_sequences(documents_df.documents_cleaned)
tokenized_paded_documents = pad_sequences(tokenized_documents, maxlen=64, padding='post')
vocab_size = len(tokenizer.word_index) + 1

# Reading Glove word embeddings into a dictionary with "word" as key and values as word vectors
embeddings_index = dict()
with open('/content/drive/MyDrive/glove.6B/glove.6B.100d.txt') as file:
    for line in file:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

# Creating embedding matrix, every row is a vector representation from the vocabulary indexed by the tokenizer index.
embedding_matrix = np.zeros((vocab_size, 100))
for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

# Creating TF-IDF vectors
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectors = tfidf_vectorizer.fit_transform(documents_df.documents_cleaned)

# Extracting feature names (words) from the TF-IDF vectorizer vocabulary
words = tfidf_vectorizer.get_feature_names_out()

# Calculating average of word vectors of a document weighted by tf-idf
document_embeddings = np.zeros((len(tokenized_paded_documents), 100))

# Instead of creating document-word embeddings, directly creating document embeddings
for i in range(documents_df.shape[0]):
    for j in range(len(words)):
        if words[j] in tokenizer.word_index:
            document_embeddings[i] += embedding_matrix[tokenizer.word_index[words[j]]] * tfidf_vectors[i, j]

pairwise_similarities = cosine_similarity(document_embeddings)
pairwise_differences = euclidean_distances(document_embeddings)
```
